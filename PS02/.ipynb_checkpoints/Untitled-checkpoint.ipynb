```python
import pandas as pd
import numpy as np 
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python


df=pd.read_csv('UpdatedStudentsPerformance.csv')
```

```python
df.head(34)
```

```python
#filling the reading score null values with mean values
newDFMean=df['reading score'].fillna(df['reading score'].mean())


# In[125]:


newDFMean.isna().sum()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['reading score'],axis=1)
```

```python
df=pd.concat([df,newDFMean],axis=1)
```

```python
df
```

```python


newDFMedian=df['math score'].fillna(df['math score'].median())


# In[133]:


# In[134]:


df=df.drop(['math score'],axis=1)

df=pd.concat([newDFMedian,df],axis=1)



newDFMode=df['writing score'].fillna(df['writing score'].mode()[0])


# In[141]:
# In[142]:


df=df.drop(['writing score'],axis=1)


# In[143]:


df=pd.concat([newDFMode,df],axis=1)
```

```python


labelencoder=preprocessing.LabelEncoder()


# In[158]:


df['gender']=labelencoder.fit_transform(df['gender'])
df['lunch']=labelencoder.fit_transform(df['lunch'])
df['race/ethnicity']=labelencoder.fit_transform(df['race/ethnicity'])
df['parental level of education']=labelencoder.fit_transform(df['parental level of education'])
df['test preparation course']=labelencoder.fit_transform(df['test preparation course'])
```
